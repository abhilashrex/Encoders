```python
import numpy as np
import pandas as pd
import calendar
from datetime import datetime

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# Loading data
data = pd.read_csv('sales.csv')

# Converting the Datetime datatype  to datetime
data['OrderDate'] = pd.to_datetime(data['OrderDate'])

# creating new columns(hour,date,weekday,month)
data['date'] = data['OrderDate'].dt.date
data["weekday"] = data['OrderDate'].dt.strftime('%A')
data["month"] = data['OrderDate'].dt.month_name()
data['year'] = data['OrderDate'].dt.year

# Categorical boolean mask
categorical_feature_mask = data.dtypes==object

#filter categorical columns using mask and turn it into a list

categorical_cols = data.columns[categorical_feature_mask].tolist()

# Although date is an object, it is been added when converting datetime to datetime obejct. so let us remove that from list.
categorical_cols.remove('date')
categorical_cols.remove("Total")
print(categorical_cols)
```
